```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('PlayerData.csv').drop(columns=['Unnamed: 0'])
data.head()
```

```python
# All players before 2001 have 'NaN' for 'pct_from_3' - Drop em
data = data[data.Year >= 2001]
data.head()
```

```python
dict_list = []
for player in data.Player.unique():

    temp_df = data[data['Player']==player]

    player_dict = dict(zip(temp_df.iloc[0].keys(), temp_df.iloc[0]))

    if temp_df.shape[0] >= 5:
        player_dict['Salary5'] = temp_df.iloc[4].Salary
        dict_list.append(player_dict)

# Make New DataFrame
new_df = pd.DataFrame(dict_list)
new_df.head()
```
